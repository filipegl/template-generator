```python
import pandas as pd
```

```python
imdb_df = pd.read_csv('./data/imdb_sampled/data-100samples.csv')
imdb_df.head()
```

```python
df = pd.read_clipboard(usecols=["template_index", "label", "original_text", "masked_text", "template_text"])
df
```

```python
def find_original_text(sentence):
    for review in imdb_df["text"].items():
        if sentence in review[1]:
            return (review)

df.insert(0, "original_review", df.original_text.map(find_original_text))
```

```python
df.to_csv("original_reviews_with_sentences/random_original_reviews.csv", index=False)
```
