```python
import pandas as pd
```

```python
for approach in ["approach1.xlsx", "approach2.xlsx", "approach3.xlsx", "approach4.xlsx", "approach5.xlsx", "random.xlsx"]:
    dfs_ann1 = pd.read_excel(f"../analise-amostragem-imdb-SRS/data_formated/data_with_templates/data_annotated/{approach}", sheet_name=None)
    dfs_ann1_divergent = pd.read_excel(f"ann1/revised/{approach}", sheet_name=None)
    # dfs_ann2 = pd.read_excel(f"../analise-amostragem-imdb-SRS/data_formated_label_adson/{approach}", sheet_name=None)
    # dfs_ann2_divergent = pd.read_excel(f"ann2/{approach}", sheet_name=None)
    
    writer_ann1 = pd.ExcelWriter(f"ann1/convergent_and_divergent/{approach}")
    # writer_ann2 = pd.ExcelWriter(f"ann2/convergent_and_divergent/{approach}")
    
    for t in dfs_ann1:
        if t != "templates":
            for i, row in dfs_ann1_divergent[t].set_index("Unnamed: 0").iterrows():
                dfs_ann1[t].loc[i] = row
            # for i, row in dfs_ann2_divergent[t].set_index("Unnamed: 0").iterrows():
            #     dfs_ann2[t].loc[i] = row
        dfs_ann1[t].to_excel(writer_ann1, sheet_name=t, index=False)
    
    writer_ann1.close()
    # writer_ann2.close()
```
